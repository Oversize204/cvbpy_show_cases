````python
import os
import numpy as np
import pyvista as pv
import panel as pn
import cvb

# we need a virtual frame buffer
pv.start_xvfb()

pn.extension('vtk', sizing_mode="stretch_width")

cvb_pc = cvb.PointCloudFactory.load("CanLid_Part.ply", cvb.PointCloudFlags.Float | cvb.PointCloudFlags.XYZ)
np_pc = cvb.as_array(cvb_pc)
point_cloud = pv.PolyData(np_pc)
point_cloud['point_color'] = point_cloud.points[:, 2]  # just use z coordinate
point_cloud.plot(notebook=True, jupyter_backend='panel', scalars='point_color')
````
